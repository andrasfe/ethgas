```python
from google.cloud import bigquery
import pandas as pd
from bq_helper import BigQueryHelper

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import math
from datetime import datetime
```

```python
bq_assistant = BigQueryHelper("bigquery-public-data", "crypto_ethereum")
```

```python
def query_and_save(query, val_column, file_name):
    df = bq_assistant.query_to_pandas_safe(query, max_gb_scanned=100)
    df.reset_index(drop=True, inplace=True)
    df.columns = ['Date', val_column]
    df['Date'] = df['Date'] + ':00'
    df=df.sort_values(by=['Date'],ascending=True)
    df.to_csv(file_name, index=False)
```

```python
query = """
SELECT 
 SUBSTR(STRING(block_timestamp), 1, 16), AVG(gas_price)
 
FROM
  `bigquery-public-data.crypto_ethereum.transactions` AS transactions
WHERE DATE(transactions.block_timestamp) BETWEEN DATE(2021,1,1) AND DATE(2021,10,18)
GROUP BY SUBSTR(STRING(block_timestamp), 1, 16)
"""
query_and_save(query, 'Price', 'transactions.csv')
```

```python
query = """
SELECT 
 SUBSTR(STRING(timestamp), 1, 16), SUM(transaction_count)
 
FROM
  `bigquery-public-data.crypto_ethereum.blocks` AS blocks
WHERE DATE(timestamp) BETWEEN DATE(2021,1,1) AND DATE(2021,10,18)
GROUP BY SUBSTR(STRING(timestamp), 1, 16)
"""

query_and_save(query, 'Count', 'blocks.csv')
```
